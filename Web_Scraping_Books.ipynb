## Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://books.toscrape.com/"
response = requests.get(url)
print(response.status_code)
```

```python
print(response.text[:500])
```

## Convert HTML into a Searchable Object

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
books = soup.find_all("article", class_="product_pod")

print(len(books))
```

## Look at the First Book

```python
books[0]
```

```python
title = books[0].h3.a["title"]
print(title)
```

## Extract the Price

```python
price = books[0].find("p", class_="price_color").text
print(price)
```

## Extract the Rating

```python
rating = books[0].find("p")["class"][1]
print(rating)
```

## Extract Availability

```python
availability = books[0].find("p", class_="instock availability").text.strip()
print(availability)
```

## Extract the Product URL

```python
link = books[0].h3.a["href"]
print(link)
product_url = "https://books.toscrape.com/" + link
print(product_url)
```

## Print Everything Together

```python
print("Title:", title)
print("Price:", price)
print("Rating:", rating)
print("Availability:", availability)
print("Product URL:", product_url)
```

## Create an Empty List

```python
book_data =[]
```

## Scrap All 20 Books

```python
book_data = []

for book in books:

    title = book.h3.a["title"]

    price = book.find("p", class_="price_color").text

    rating = book.find("p")["class"][1]

    availability = book.find("p", class_="instock availability").text.strip()

    link = book.h3.a["href"]
    product_url = "https://books.toscrape.com/" + link

    book_data.append({
        "Title": title,
        "Price": price,
        "Rating": rating,
        "Availability": availability,
        "Product URL": product_url
    })
```

## Check the Data

```python
book_data
```

## Convert to a DataFrame

```python
df = pd.DataFrame(book_data)
df
```

## Save It as a CSV File

```python
df.to_csv("books.csv", index=False)
print("Dataset saved successfully!")
```

## Scrap All 50 Books

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

book_data = []

BASE_URL = "https://books.toscrape.com/"

for page in range(1, 51):

    if page == 1:
        url = BASE_URL
    else:
        url = BASE_URL + f"catalogue/page-{page}.html"

    print(f"Scraping Page {page}...")

    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    books = soup.find_all("article", class_="product_pod")

    for book in books:

        title = book.h3.a["title"]

        price = book.find("p", class_="price_color").text

        rating = book.find("p")["class"][1]

        availability = book.find(
            "p",
            class_="instock availability"
        ).text.strip()

        link = book.h3.a["href"]

        if page == 1:
            product_url = BASE_URL + link
        else:
            product_url = BASE_URL + "catalogue/" + link.replace("../", "")

        book_data.append({
            "Title": title,
            "Price": price,
            "Rating": rating,
            "Availability": availability,
            "Product URL": product_url
        })

print("Scraping Completed!")
```

## Check the Number of Books

```python
len(book_data)
```

## Convert to DataFrame

```python
df=pd.DataFrame(book_data)
df.head()
```
